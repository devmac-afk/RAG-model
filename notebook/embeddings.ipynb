```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
import os
```

```python
import google.generativeai as genai

# Configure with your GOOGLE_API_KEY
genai.configure(api_key= os.environ['GOOGLE_API_KEY'])

print('Available models:')
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

```python
# Prompt for image description
prompt_template_vision = """
You are an assistant tasked with summarizing tables and text.
Give a concise summary of the table or text.

Respond only with the summary, no additionnal comment.
Do not start your message by saying "Here is a summary" or anything like that.
Just give the summary as it is.

Table or text chunk: {element}
"""
messages_vision = [
    (
        "user",
        [
            {"type": "text", "text": prompt_template_vision},
            {
                "type": "image_url",
                "image_url": {"url": "data:image/jpeg;base64,{image}"},
            },
        ],
    )
]

prompt_vision = ChatPromptTemplate.from_messages(messages_vision)

# Initialize the multimodal model (Gemini Pro Vision)
# Ensure GOOGLE_API_KEY is set in your environment or Colab secrets
model_vision = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Create the summarization chain for images
image_summarization_chain = prompt_vision | model_vision | StrOutputParser()
```
